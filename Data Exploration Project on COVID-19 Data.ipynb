# Initial Data Overview
It provides an overview of the COVID-19 dataset by loading it and displaying the first few rows to understand its structure and contents.

```python
data = pd.read_csv("dcov.csv")
print("Data Overview")
print(data.head())
```

## Missing Values
It identifies and prints the number of missing values in the dataset. Then, it removes rows with missing values and displays the count of remaining data points.

```python
print("Missing values:")
print(data.isnull().sum())
data = data.dropna()
data.count()
```

## Data Overview

```python
# Displaying the DataFrame
print(data.head())
```

## Correlation Analysis
It visualizes the correlation matrix between numerical columns in the COVID-19 dataset using a heatmap, highlighting the strength and direction of relationships between variables.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numerical Columns')
plt.show()
```

## Sleep Hours Distribution by COVID-19 Class
It visualizes the distribution of sleep hours among different COVID-19 classes using a boxplot, providing insights into the relationship between sleep patterns and disease severity.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='sleephr', data=data, palette='pastel')
plt.title('Sleep Hours by Class')
plt.xlabel('Class')
plt.ylabel('Sleep Hours')
plt.show()
```

## Heart Rate Variation by Appetite and COVID-19 Class
It illustrates the distribution of heart rate across different levels of appetite and COVID-19 classes using a boxplot, aiding in understanding the relationship between appetite, disease severity, and heart health.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='appetite', y='heart_rate', hue='class', data=data)
plt.title('Heart Rate by Appetite and COVID-19 Class')
plt.xlabel('Appetite')
plt.ylabel('Heart Rate')
plt.show()
```

# Hexbin Map of Random Data
It displays a hexbin map of randomly generated geographical coordinates, providing a visual representation of data density and distribution across a geographical area.

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming data contains geographical coordinates
plt.figure(figsize=(10, 8))

# Generate random data for demonstration
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.hexbin(x, y, gridsize=30, cmap='viridis')
plt.colorbar(label='count in bin')
plt.title('Hexbin Map of Random Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Heart Rate Distribution by Obesity and COVID-19 Class
It showcases the distribution of heart rate across different levels of obesity and COVID-19 classes using a violin plot, offering insights into the relationship between obesity, disease severity, and heart health.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='obesity', y='heart_rate', hue='class', data=data)
plt.title('Heart Rate by Obesity and COVID-19 Class')
plt.xlabel('Obesity')
plt.ylabel('Heart Rate')
plt.show()
```

# Heart Rate Trend Over Time
It visualizes the trend of heart rate over time using a line plot, providing insights into the temporal patterns of heart health within the COVID-19 dataset.

```python
# Example: Line plot of Heart Rate over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x=data.index, y='heart_rate', data=data)
plt.title('Heart Rate over Time')
plt.xlabel('Time')
plt.ylabel('Heart Rate')
plt.show()
```

# Pairwise Relationship Exploration
It explores the pairwise relationships between selected features (Duration, Active Time, Heart Rate, Sleep Hours) and COVID-19 class using a pairplot, highlighting potential patterns and correlations between variables.

```python
sns.pairplot(data[['Duration', 'Active_time', 'heart_rate', 'sleephr', 'class']], hue='class')
plt.show()
```
